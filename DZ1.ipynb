## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

### My solution:

````python
def rotate(matrix, how):
    if how == 'right':
        return [list(row)[::-1] for row in zip(*matrix)]
    elif how == 'left':
        return [list(row) for row in zip(*matrix)][::-1]
    
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

#по часовой
for row in rotate(matrix, 'right'):
    print(row)

print()    
    
#против часовой
for row in rotate(matrix, 'left'):
    print(row)
````

## Problem 3

Write a function that takes a string as input and returns a dictionary containing the number of occurrences of each character in the string. Use method `count` of strings. Implement this function in one line using $\lambda$-function.

Example for the string 'hello, world!': {'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, ' ': 1, 'w': 1, 'r': 1, 'd': 1, '!': 1}.

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


### My solution:

````python
class Book:
    def __init__(self, title, author, year, isbn, copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies
        print(f"Book '{title}' created")

    def get_info(self):
        return f"{self.title} by {self.author} ({self.year}) | ISBN: {self.isbn} | Copies available: {self.copies}"

    def change_copies(self, amount):
        self.copies += amount
        print(f"Copies of '{self.title}' updated. Now available: {self.copies}")


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"User {name} registered")

    def borrow_book(self, book):
        if book.copies > 0 and book not in self.borrowed_books:
            book.change_copies(-1)
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'")
            return True
        print(f"Borrowing failed for '{book.title}'")
        return False

    def return_book(self, book):
        if book in self.borrowed_books:
            book.change_copies(1)
            self.borrowed_books.remove(book)
            print(f"{self.name} returned '{book.title}'")
            return True
        print(f"Return failed. '{book.title}' not borrowed by {self.name}")
        return False

    def list_borrowed_books(self):
        return [book.get_info() for book in self.borrowed_books]


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.transaction_history = []
        print("Library system created")

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)

    def register_user(self, user):
        self.users.append(user)

    def remove_user(self, user):
        if user in self.users:
            self.users.remove(user)

    def issue_book(self, user, book):
        if user.borrow_book(book):
            self.transaction_history.append(f"{user.name} borrowed '{book.title}'")
            return True
        return False

    def return_book(self, user, book):
        if user.return_book(book):
            self.transaction_history.append(f"{user.name} returned '{book.title}'")
            return True
        return False

    def search_book_by_title(self, title):
        return [book.get_info() for book in self.books if title.lower() in book.title.lower()]

    def search_user_by_name(self, name):
        return [user.name for user in self.users if name.lower() in user.name.lower()]

    def display_transaction_history(self):
        for record in self.transaction_history:
            print(record)


# Example usage
library = Library()

# Creating books
book1 = Book("банда Ван дер Линде", "Артур Морган", 1887, "12345676767", 82000000)
library.add_book(book1)

# Registering users
user1 = User("Дмитрий Воронин", "К15003")
library.register_user(user1)

# Issuing book
library.issue_book(user1, book1)

# Returning book
library.return_book(user1, book1)

# Displaying transaction history
library.display_transaction_history()
````

## Problem 6*

Let 
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

### My solution:

````python
A = 0.0
for i in range(1, 10001, 1):
    A += 1 / (i * i)
B = 0.0
for i in range(10000, 0, -1):
    B += 1 / (i * i)

print(f'A = {A}')
print(f'B = {B}')
print(f'B - A = {B - A}')
````

При вычислении A мы сначала складываем большие дроби, а потом прибавляем к ним маленькие. Из-за ограниченной точности компьютера эти мелкие числа просто теряются, когда попадают в большую сумму — он их "не замечает". А в варианте B всё наоборот: мы сначала накапливаем эти самые маленькие дроби, они образуют ощутимую сумму, и только потом добавляем к ним крупные числа. Поэтому потерь почти нет. Значит, правильнее суммировать от самых маленьких слагаемых к самым большим.
